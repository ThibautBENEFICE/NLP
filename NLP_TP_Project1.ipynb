Naive Bayes classifier

```python
lower_train = list(map(str.lower, dataset_train['text']))
len(dataset_train['text'])
```

```python
from string import punctuation
no_punct =  list(map(lambda input : input.translate(str.maketrans('', '', punctuation)), lower_train))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
vectorizer_train = CountVectorizer()
X = vectorizer_train.fit_transform(no_punct)
```

```python
X[:5].toarray()
```

```python
clf = MultinomialNB()
clf.fit(X, dataset_train['label'])
print(clf.predict(X[100]))
```

```python
#pipeline tests
lower_test = list(map(str.lower, dataset_test['text']))
no_punct_test =  list(map(lambda input : input.translate(str.maketrans('', '', punctuation)), lower_test))
X_test = vectorizer_train.transform(no_punct_test)
```
